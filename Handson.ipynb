```python
print(" Pay Check Calculator\n")
hours = float(input("Hours Worked: "))
payrate = float(input("Hourly Pay Rate: "))
grosspay = hours*payrate
print("\nGross Pay: " ,  grosspay)
taxrate = float(input("Tax Rate: "))
taxamount = grosspay*(taxrate/100)
print("Tax Amount: ", taxamount)
THP= grosspay - taxamount
print("Take Home Pay: ", THP)
```

```python
print("Travel Time Calculator\n")
miles=int(input("Enter miles: "))
mph=int(input("Enter miles per hour: "))
print("\nEstimated travel time")
hours=int(miles/mph)
print("Hours: " ,hours)
minutes=miles%mph
print("Minutes: ", minutes)
```
